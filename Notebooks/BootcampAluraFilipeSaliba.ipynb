#Caderno de Exploração dos Dados do DATASUS a partir de 2008 / Aula 01 - Bootcamp Alura

**Importando o Pandas e upando os dados.**

```python
import pandas as pd

dados = pd.read_csv('/content/A104207189_28_143_208.csv', encoding='ISO-8859-1',
                    skiprows = 3, sep=';', skipfooter = 12,
                    thousands='.', decimal=',')

pd.options.display.float_format = '{:.2f}'.format

dados.head()
```

#Plotando o primeiro gráfico com os dados carregados

```python
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


ax = dados.plot(x="Unidade da Federação", y="2021/Mar", kind="bar", figsize=(10,7))

ax.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
ax.set_facecolor('white')
plt.title("SUS: Valores Gastos com Procedimentos Hospitalares por Estado em Mar/2021", fontsize=17, fontweight="bold")
plt.show()
```

**Exibindo dados selecionados**

```python
dados[['2008/Ago', '2008/Set']].head()
```

```python
dados.mean()
```

```python
colunas_usaveis = dados.mean().index.tolist()
colunas_usaveis.insert(0, "Unidade da Federação")
colunas_usaveis
```

```python
usaveis = dados[colunas_usaveis]
usaveis.head()
```

```python
usaveis = usaveis.set_index("Unidade da Federação")
```

```python
usaveis["2019/Ago"].head()
```

```python
usaveis.loc["12 Acre"]
```

```python
usaveis.plot(figsize=(10,6))
```

```python
usaveis.T.head()
```

```python
usaveis.T.plot(figsize=(10,7))
```

```python
usaveis = usaveis.drop("Total", axis=1)
```

```python
usaveis.tail()
```

```python
usaveis.T.plot(figsize=(10,7.5))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Valor total por Ano/mês processamento segundo Unidade da Federação', fontsize=14, fontweight="bold")
```

```python
EstadosSudeste_EParana = usaveis.loc[['31 Minas Gerais', '35 São Paulo', '41 Paraná', '33 Rio de Janeiro', '32 Espírito Santo']]
EstadosSudeste_EParana
```

```python
EstadosSudeste_EParana.T.plot(figsize=(10,7.5))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
plt.title('Valor total por Ano/mês processamento estados selecionados', fontsize=14, fontweight="bold")
plt.show()
```
